- Task: Read your `maha.txt` file. Use `spaCy` to extract all Named Entities from the text.
- Output: Use Python's `collections.Counter` to map the categorical frequency distribution of entity types. Print a clean dashboard showing the dynamic tally counts (e.g., `ORG: X times`, `PERSON: Y times`, `GPE: Z times`). Separately isolate and print all unique tokens labeled as `PERSON`.

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
with open("maha.txt","r") as f:
  text = f.read()
```

```python
doc = nlp(text)
```

```python
# NER in first 50 tokens
for ent in doc[:50].ents:
  print(ent.text , ent.label_)
```

```python
# printing entire article and NER
from spacy import displacy
```

```python
# for first 150 tokens
displacy.render(doc[:150], style="ent")
```
