- Task: Use NLTK or spaCy to read your `friends.txt` file and build a cleaning routine:
    
    1. Tokenize into sentences and words.
    2. Strip out all standard English stop words and punctuation.
    3. Run the filtered words through both the Porter Stemmer or WordNet Lemmatizer.

```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
```

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
file = "friends.txt"
with open(file, 'r') as f:
    text = f.read()
```

```python
clean = re.sub(r'[^\w\s]', '', text)

tokens = word_tokenize(clean)


tokens = [token.lower() for token in tokens]
print(f"Before removing stop words : {len(tokens)}")

sw = set(stopwords.words('english'))

tokens = [token for token in tokens if token not in sw ]
print(f"After removing stop words : {len(tokens)}")
```

```python
ps = PorterStemmer()
stem_tokens = [ps.stem(word) for word in tokens ]


# Printing 5 words - lemmatized version
for i , token in enumerate(tokens[:10]):
    print(f"{token} : {stem_tokens[i]}")
```
