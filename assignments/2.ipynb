🏷️ ASSIGNMENT 2: Syntactic Tagging (Using `2e.txt`)

- Task: Read your scraped `2e.txt` file. Use `spaCy`'s English model to parse the text and tag grammatical categories.
- Output: Loop through tokens and map out each word alongside its Universal POS tag (`token.pos_`) and detailed tag (`token.tag_`). Identify contextual areas where the model tripped up due to lexical ambiguity.

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
with open("2e.txt", "r") as f:
  text = f.read()
```

```python
doc = nlp(text)
```

```python
# printing first 5 pos tags
for token in doc[:5]:
  print()
  print(f"Word : {token}")
  print(f"POS tag : {token.pos_, token.tag_}")
  print("*---*---*---"*2)
```

### Extra : Lemmatization

```python
for token in doc[:5]:
  print()
  print(f"{token} ---> {token.lemma_}")
```
